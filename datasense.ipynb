```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.pipeline import Pipeline
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
#from sklearn.linear_model import LinearSVM
from sklearn.metrics import classification_report
import joblib
```

```python
"""vectorizer= TfidfVectorizer(stop_words='english')
tfidf = vectorizer.fit_transform(df['sms_text'])"""
```

```python
x = dum_df[['masked_sms_text']]
y = dum_df['encoded_category']
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state= 42,stratify=y)
```

```python
x_train
```

```python
transformer = ColumnTransformer([
    ('text',TfidfVectorizer(stop_words='english'),'masked_sms_text')
]
)
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
"""params = {
    'n_estimators':[50,100,200],
    #'max_leaf_nodes':[,2,3,4],
    'min_samples_leaf': [5,10,20],
    'min_samples_split':[10,20],
    'max_depth':[10,20,30,40]
}"""
params = {
    # 1. Bring back 'None' or very deep trees.
    # Text usually needs depth > 50 to capture vocabulary.
    'max_depth': [None, 100, 200],

    # 2. Lower the leaf requirement.
    # '1' allows memorization. '2' or '4' cuts just the extreme noise.
    # '10' (from before) was clearly too high for this dataset.
    'min_samples_leaf': [1, 2, 4],

    # 3. Standard split requirements
    'min_samples_split': [2, 5, 10],

    # 4. Increase trees to stabilize variance
    'n_estimators': [100, 200]
}
```

```python
grid_search = GridSearchCV(RandomForestClassifier(),param_grid=params,scoring = 'accuracy',n_jobs=-1,cv=5)
```

```python
pipeline = Pipeline([
    ('processor',transformer),
    ('model',LogisticRegression())
])
```

```python
pipeline.fit(x_train,y_train)
```

```python
#pipeline.named_steps['model'].best_params_
```

```python
#pipeline.named_steps['model'].best_score_
```

```python
print('test score ',pipeline.score(x_test,y_test))
print('train score ',pipeline.score(x_train,y_train))
```

```python
pred = pipeline.predict(x_test)
```

```python
print(classification_report(y_test,pred))
```

```python
joblib.dump(pipeline, 'full_classification_pipeline.joblib')
```
